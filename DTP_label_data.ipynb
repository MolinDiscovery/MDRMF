```python
import pandas as pd
import numpy as np

data = pd.read_csv('DTP_data_final_fix1.csv', index_col=0)
data
```

```python
docking_scores = {
    517865: -5.22,
    43359: -30.21,
    19139: -0.21,
    16612: -6.82,
    69356: -55.12,
    109327: -7.21,
    120688: -0.51,
    303769: -42.12,
    320852: -72.53,
    133656: -0.12,
}

data['docking_score'] = data['NSC'].map(docking_scores)
data
```

```python
NSC_list = list(docking_scores.keys())
NSC_indexes = data['NSC'].isin(NSC_list)
data_with_score = data[NSC_indexes]
data_with_score
```

```python
# Make dataset smaller for developing purposes

data_len = 200

if len(data_with_score) >= data_len:
    short_data = data_with_score.head(data_len)
else:
    remaining_rows = data_len - len(data_with_score)
    data_without_score = data[data['docking_score'].isna()].head(remaining_rows)
    short_data = pd.concat([data_with_score, data_without_score])

short_data = short_data.sort_values(by=['NSC'])
short_data
```
