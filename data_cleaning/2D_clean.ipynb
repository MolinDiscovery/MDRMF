```python
from rdkit import Chem

def canonicalize_smiles(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        return Chem.MolToSmiles(mol, isomericSmiles=True)
    else:
        return None
    
df_10K['Canonical_SMILES'] = df_10K['SMILES'].apply(canonicalize_smiles)
canonical_smiles_to_bits = {canonicalize_smiles(smiles): bits for smiles, bits in smiles_to_bits.items()}

df_10K['bits'] = df_10K['Canonical_SMILES'].map(canonical_smiles_to_bits)
```

Downloaded the original datafile found on slack and matched docking scores through the title column. After that added the i_i_glide_lignum through which the original SMILES could be mapped onto a final dataframe.

```python
df_Z_ids = pd.read_csv('Z-ids.csv')
```

```python
title_to_bits = {row['title']: row[2:].values.astype('int8') for _, row in df.iterrows()}
```

```python
# Create a new dataframe from df_Z_ids with 'title' and 'r_i_docking_score'
merged_df_Z_ids = df_Z_ids[['title', 'r_i_docking_score']]

# Map the bits from the dictionary to the new dataframe based on the 'title'
merged_df_Z_ids['bits'] = merged_df_Z_ids['title'].map(title_to_bits)

# Drop rows where the 'bits' column is NaN (indicating no match in the dictionary)
merged_df_Z_ids.dropna(subset=['bits'], inplace=True)

merged_df_Z_ids.head()
```

```python
merged_df_Z_ids
```

```python
merged_df_Z_ids['i_i_glide_lignum'] = df_Z_ids['i_i_glide_lignum']
```

```python
merged_df_Z_ids
```

```python
# Create a mapping of i_i_glide_lignum to SMILES from df_10K
lignum_to_smiles = df_10K.set_index('i_i_glide_lignum')['SMILES'].to_dict()

# Map SMILES to merged_df_Z_ids using i_i_glide_lignum
merged_df_Z_ids['SMILES'] = merged_df_Z_ids['i_i_glide_lignum'].map(lignum_to_smiles)

# Display the first few rows of the updated dataframe
merged_df_Z_ids.head()
```

Ensuring the SMILES are correct (I don't remember where the old ones came from)

```python
# Load the HLL10k.csv file
df_HLL10k = pd.read_csv('HLL10k.csv')

# Create a mapping of ids to SMILES from df_HLL10k
ids_to_smiles_HLL = df_HLL10k.set_index('ids')['SMILES'].to_dict()

# Replace SMILES in merged_df_Z_ids using the mapping
merged_df_Z_ids['SMILES'] = merged_df_Z_ids['title'].map(ids_to_smiles_HLL)

# Display the first few rows of the updated dataframe
merged_df_Z_ids.head()
```

```python
# Export dataset
merged_df_Z_ids.to_csv("2D.csv")
```

```python
# Make a dataset, that the model can work with
from MDRMF import Dataset
import numpy as np

X = merged_df_Z_ids['bits']
X = np.vstack(X)
y = merged_df_Z_ids['r_i_docking_score']
ids = merged_df_Z_ids = merged_df_Z_ids['SMILES']

dataset_2D = Dataset(X, y, ids)
dataset_2D.save("2D.pkl")
```
