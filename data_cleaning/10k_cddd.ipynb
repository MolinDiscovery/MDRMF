```python
from MDRMF import Dataset
from MDRMF import MoleculeLoader
from MDRMF import Featurizer
import numpy as np
import pandas as pd
```

```python
data = MoleculeLoader('10K.csv', 'SMILES', 'r_i_docking_score').df
#data = data.iloc[:-5000]

feat = Featurizer(data)
features = feat.featurize("morgan", radius=2, nBits=1024)

X = features
y = data['r_i_docking_score']
ids = data['SMILES']

dataset = Dataset(X=X, y=y, ids=ids)
```

```python
iteration = 0

for smile in data['smiles']:
    count = np.count_nonzero(dataset.ids == smile)
    if count != 1:
        print(f"Problem in iteration:\t{iteration}")
        print(f"SMILE in cddd :\t\t{smile}")
        print(f"SMILE in dataset:\t{dataset.ids[iteration]}")
        print(f"Count\t\t\t{count}")
        print("")
    
    if count == 0:
        print("No matches for {smile}")

    iteration = iteration + 1
```
